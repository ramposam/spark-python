```python
import pandas as pd
```

```python
empschema =['empno', 'ename', 'job', 'mgr', 'hiredate', 'sal', 'comm', 'deptno']
```

```python
def convert_na_zero(col):
    if col in ("Nan","Na","Null","null","NAN"):
        return float(0)
    return col
emp = pd.read_csv("data/emp.txt",sep=":",names=empschema,converters={"comm":convert_na_zero})
emp
```

```python
emp["comm"] = emp["comm"].fillna(0).astype(float)
emp.set_index("empno")
emp
```

```python
emp_g = emp.groupby("deptno")
```

```python
emp_g.describe()
```

```python
emp_g.get_group(10)
```

```python
emp_g.max()
```

```python
emp_g.mean()
```

```python
emp_g.max().filter(["sal","comm"]).rename(columns={"sal":"max_sal","comm":"max_comm"})
```

```python
emp_g.sum().filter(["sal","comm"]).rename(columns={"sal":"total_sal","comm":"total_comm"})
```

```python
emp_g.sum().filter(["sal","comm"]).rename(columns={"sal":"total_sal","comm":"total_comm"}).to_csv("/home/rposam2020/output/emp_sum.csv")
```

```python
deptschema = ['deptno', 'dname', 'loc']
dept = pd.read_csv("data/dept.csv",names=deptschema)
dept.set_index("deptno")
dept
```

```python
emp_dept = pd.merge(left=emp,right=dept,left_on="deptno",right_on="deptno")
emp_dept
```

```python
emp_dept_left_join = pd.merge(left=emp,right=dept,how="left",left_on="deptno",right_on="deptno")
emp_dept_left_join
```

```python
emp_dept_right_join = pd.merge(left=emp,right=dept,how="right",left_on="deptno",right_on="deptno")
emp_dept_right_join.replace({"NaN":0})
emp_dept_right_join
```

```python
emp_dept_full_join = pd.merge(left=emp,right=dept,how="outer",left_on="deptno",right_on="deptno")
emp_dept_full_join
```

```python
emp_dept_sel_cols = pd.merge(left=emp[["empno","ename","job","sal","deptno"]],right=dept,how="outer",left_on="deptno",right_on="deptno")
emp_dept_sel_cols
```
