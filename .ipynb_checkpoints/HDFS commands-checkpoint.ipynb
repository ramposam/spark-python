```python
%%sh
##HDFS commands:
##List all files/folders in home directory of user
hadoop fs -ls 

##To List all files/folders and it's underlying files/directories
hadoop fs -ls -R 

##To sort files by time ascending
hadoop fs -ls -t

##To sort files by time descending
hadoop fs -ls -t -r

##To sort files by size ascending
hadoop fs -ls -S -r /public/yelp-dataset

##To sort files by size descending
hadoop fs -ls -S /public/yelp-dataset

##To display file size in Kb/MB/GB in descending order
hadoop fs -ls -S -h /public/yelp-dataset

##To sort files path
hadoop fs -ls -C /public/yelp-dataset

##To search file/folder and it's underlying files/folders with matching name
hadoop fs -ls -R /public|grep order

##for userspace
hadoop fs -ls /user|grep rposam
hadoop fs -ls /user|grep `whoami`

##to create directories
hadoop fs -mkdir /user/rposam2020/test

##if file already exists then
hadoop fs -mkdir -p /user/rposam2020/test

##to create multiple directories
hadoop fs -mkdir -p /user/rposam2020/highlevel/lowlevel

##To delete underlying files/directories
hadoop fs -rm -R /user/rposam2020/highlevel

##to remove directory this throws error if the directory is non-empty
hadoop fs -rmdir  /user/rposam2020/highlevel

##To skip that error display
hadoop fs -rmdir --ignore-fail-on-non-empty  /user/rposam2020/highlevel

##Copy Files/folders to HDFS
hadoop fs -mkdir -p /user/rposam2020/retail_db
hadoop fs -ls /user/rposam2020/

hadoop fs -copyFromLocal /data/retail_db/orders /user/rposam2020/retail_db/
hadoop fs -put /data/retail_db/orders /user/rposam2020/retail_db/

hadoop fs -ls /user/rposam2020/retail_db/orders
hadoop fs -rm -R /user/rposam2020/retail_db/orders

##To overwrite files/directories if exists
hadoop fs -put -f /data/retail_db/orders /user/rposam2020/retail_db/

##Permissions 
##Owener-group-others Permissions
##-rw-r--r--   2 rposam2020 hdfs    2999944 2020-02-09 03:57 /user/rposam2020/retail_db/orders/part-00000

##Copy Files/folders to Local
hadoop fs -copyToLocal /user/rposam2020/data/ /home/rposam2020/ 
hadoop fs -get /user/rposam2020/data/ /home/rposam2020/ 

##View Files/folders
view /data/emp.csv
head /data/emp.csv
tail /data/emp.csv

##hdfs doesn't support tail command if you want to see first few lines of header use following command
hadoop fs -cat /user/rposam2020/data/emp*|more


##Copy/Move HDFS to HDFS
hadoop fs -ls /user/rposam2020/retail_db/

hadoop fs -cp /public/cards /user/rposam2020/retail_db/.
hadoop fs -mv /user/rposam2020/retail_db/cards /user/rposam2020/retail_db/.

##File system size and available storage info including file system path
hadoop fs -df -h /user/rposam2020/

##For Directory sizes
hadoop fs -du -h /user/rposam2020/retail_db/decks
hadoop fs -du -h /public

##For Directory total size
hadoop fs -du -s -h /public

##To copy file using replication factor
hadoop fs -rm -R /user/rposam2020/cards/

hadoop fs -Ddfs.replication=3 -put  /data/cards/largedeck.txt /user/rposam2020/largedeck.txt 
hdfs fsck /user/rposam2020 -files
##for Blocks
hdfs fsck /user/rposam2020 -files -blocks
##for locations
hdfs fsck /user/rposam2020 -files -blocks -locations

##To get CPU information:
Lscpu
```
